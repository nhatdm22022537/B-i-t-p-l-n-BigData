```python
import pandas as pd 

df = pd.read_csv("modified_movies.csv")
```

```python
df
```

```python
movie_ids = {movie: idx + 1 for idx, movie in enumerate(df['Movie_Name'].unique())}
```

```python
df['Movie_ID'] = df['Movie_Name'].map(movie_ids)
```

```python
df
```

```python
df.to_csv('movies_with_ids.csv', index=False)
```

```python
columns_to_drop = ['Unnamed: 0', 'Unnamed: 0.1']
df.drop(columns_to_drop, axis=1, inplace=True)
```
